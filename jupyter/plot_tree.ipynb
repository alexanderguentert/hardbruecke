```python
import pickle
from matplotlib import pyplot as plt
from sklearn.tree import DecisionTreeRegressor
from sklearn import tree
import pybaobabdt
#from scipy.io import arff
```

```python
XList = [
    'year',
    'hour',
    'weekday',
    'minute',
    'month',
    'direction_cat',
    'name_cat',
]
```

```python
# load model
filename_model = '../models/DecisionTreeRegressor.sav'
regressor = pickle.load(open(filename_model, 'rb'))
```

```python
fig = plt.figure(figsize=(100,40))
_ = tree.plot_tree(regressor, 
                   feature_names=XList, 
                   max_depth=4,
                   rounded=True,
                   filled=True)
```

```python
#fig.savefig("../image/decistion_tree.png")
```

```python
ax = pybaobabdt.drawTree(regressor, size=10, dpi=72, features=XList)
```

```python
ax.get_figure().savefig('../image/tree_pretty.png', format='png', dpi=300, transparent=True)
```
